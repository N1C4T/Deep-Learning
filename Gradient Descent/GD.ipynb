```python
import pandas as pd
import numpy as np 
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('insurance_data.csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['age', 'affordibility']], df['bought_insurance'], test_size = 0.2, random_state = 0 )
```

```python
y_train.head()
```

```python
X_train_scaled = X_train.copy()
X_train_scaled['age'] = X_train_scaled['age']/100

X_test_scaled = X_test.copy()
X_test_scaled['age'] = X_test_scaled['age']/100

X_train_scaled.head()
```

```python
model = keras.Sequential([
    keras.layers.Dense(1,input_shape = (2,), activation = 'sigmoid', kernel_initializer = 'ones', bias_initializer = 'zeros')
])

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train_scaled, y_train, epochs = 5)
```

```python
class myNN:
    def _init__(self):
        self.w1 = 1
        self.w2 = 1
        self.bias = 0

    def fit(self, X, y, epochs, treshold):
        self.w1, self.w2, self.bias = self.gradient_descent(X['age'], X['affordibility'], y, epochs, treshold)

    def predict(self, X):
        weighted_sum = self.w1*X['age'] + self.w2*X['affordibility'] + self.bias
        return self.np_sigmoid(weighted_sum)
    
    def np_sigmoid(self, Y):
        return 1/(1 + np.exp(-Y))

    def log_loss(self, y_true, y_predicted):
        epsilon = 1e-15
        y_predicted_new = [max(i, epsilon) for i in y_predicted]
        y_predicted_new = [min(i, 1 - epsilon) for i in y_predicted_new]
        y_predicted_new = np.array(y_predicted_new)
        return -np.mean(y_true*np.log(y_predicted) + (1 - y_true)*np.log(1 - y_predicted))
        
    def gradient_descent(self, x1, x2, y_true, epochs, treshold):
        w1 = w2 = 1
        bias = 0
        n = len(x1)
        rate = 0.5
        
        for i in range(epochs):
            weighted_sum = w1*x1 + w2*x2 + bias
            y_predicted = self.np_sigmoid(weighted_sum)
            loss = self.log_loss(y_true, y_predicted)
            
            w1d = (1/n)*(np.dot(np.transpose(x1), y_predicted - y_true))
            w2d = 1/n*(np.dot(np.transpose(x2), y_predicted - y_true))
            biasd = np.mean(y_predicted - y_true)
    
            w1 = w1 - rate*w1d
            w2 = w2 - rate*w2d
            bias = bias - rate*biasd

            if i%50 == 0:
                print(f'epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')
            
            if loss<= treshold:
                break
        print(f'epoch:{i}, w1:{w1}, w2:{w2}, bias:{bias}, loss:{loss}')
        
        return w1, w2, bias
            
```

```python
my_model = myNN()
my_model.fit(X_train_scaled, y_train, epochs = 500, treshold = 0.4331)
```
