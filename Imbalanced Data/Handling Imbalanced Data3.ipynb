```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Churn_Modelling.csv')
df.sample(5)
```

```python
df1 = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis = 1)
df1.sample(5)
```

```python
def col_val(df):
    for col in df:
        print(f'{col}: {df[col].unique()}')
col_val(df1)
```

```python
df1['Gender'] = df1['Gender'].replace({'Female': 0, 'Male': 1})
df2 = pd.get_dummies(data = df1, columns = ['Geography']).astype(float)
df2.sample(5)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
cols=['CreditScore', 'Age', 'Tenure','Balance', 'NumOfProducts', 'EstimatedSalary']
df2[cols] = scaler.fit_transform(df2[cols])
df2.dtypes
```

```python
X = df2.drop('Exited', axis = 1)
y = df2['Exited']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
X_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import classification_report 

def deep(X_train, X_test, y_train, y_test):
    model = keras.Sequential([
        keras.layers.Dense(10, input_dim = 12, activation = 'relu'),
        keras.layers.Dense(1, activation = 'sigmoid')
    ])
    model.compile(
        loss = 'binary_crossentropy',
        optimizer = 'adam',
        metrics = ['accuracy']
    )
    model.fit(X_train, y_train, epochs = 20)

    print(model.evaluate(X_test, y_test))

    y_preds = model.predict(X_test)
    y_preds = np.round(y_preds)

    print('Classification report: ', classification_report(y_test, y_preds))

    return y_preds

y_preds = deep(X_train, X_test, y_train, y_test)
```
